### 조건문과 반복문
    - 조건문 작성: if, if else, if elif else
    - 반복문 작성: for, while, list conprehension

````python
# if문 : 조건이 True일 경우 실행
if True:
    print("True일 때 출력")

if len(ls) < 3:
    print("False라서 출력 안됨")
else:
    print("조건이 False라서 else가 실행됨")
````

````python
# 반복문 for: 다른 언어와 마찬가지로 반복이 가능한(iterable) 데이터 또는 루프의 수에 따라 반복 시행
for l in ls: # ls는 iterable한 데이터 타입
    print(l)
````

````python
for t in "이것도 반복해서 출력해보세요":
    print(t)
````

````python
# dictionary는 데이터는 존재하나 순서가 없으므로 반복이 불가능하지만 파이썬에서 자체적으로 입력순으로 출력
for k,v in dict.items():
    print(k, " ", v)
````

````python
# 반복문 while: 조건이 True인 경우 무한히 반복 시행. 따라서 적절한 탈출 조건을 생성해야함
i = 0
while True:
    print(ls[i])

    i+=1 # 변수 할당

    if i == len(ls): # 탈출 조건문
        break # 조건이 True일 경우 루프 종료
````

````python
# 반복문(루프문)의 경우 조건문과 결합하여 특정 루프를 건너뛸 수 있음
for t in "1234567890":
    if int(t) == 4: # 해당 조건이 true일 경우 반복문 아래 실행문이 진행되지 않고 루프를 건너뜀
        continue
    print(t)
````

#### 전역변수, 지역변수
    - 활성화된 세션내의 변수는 어디서든 사용가능
    - 지역변수: 함수 내부의 변수는 함수 안에서만 사용되고 외부로 뽑아낼 수 없음
    - 전역변수: 외부변수를 함수 내부에서도 사용하고 외부에도 그대로 반영할 수 있도록 선언
        + global 선언

````python
num = 619

def today():
    print(num)

today()
print(f"기본 변수: num={num}")
````

````python
num = 619
def tommorow():
    num = 620
    print(num)

tommorow()
print(f"처음 선언한 변수 num={num}")
````

````python
num = 619
print(f"처음 선언한 num={num}")

def day_after_tommorrow():
    global num

    num += 2
    print(num)

day_after_tommorrow()
print(f"처음 선언한 num은 바뀌어서 {num}로 출력되었습니다.")
````

### 클래스
    - 파이썬에서 모듈화하여 사용하는 대표적인 형태
    - 클래스 내부에 변수, 함수 등을 선언하여 사용
    - 클래스 인스턴스화를 통해 객체지향 프로그래밍

````python
class Study:
    def __init__(self): # 생성자
        self.parti_ls = []
        self.student = []

        print("스터디 그룹이 생성되었습니다.")

    def join_study(self, name, age, cls):
        self.student.append(name)
        self.student.append(int(age))
        self.student.append(cls)

        self.parti_ls.append(self.student)

        print(f"스터디 그룹에 {self.parti_ls[-1]}가 참여하였습니다.")
````

````python
ml_study = Study()
ml_study.join_study("딥노이드", 37, "머신러닝")
````

#### 클래스 상속
    - 부모 클래스의 내용을 그대로 받아 사용(상속)
    - 필요에 따라 부모 클래스의 함수나 변수를 변경하여 사용

````python
class GroupStudy(Study):
    def __init__(self):
        super().__init__()
        self.group_ls = []

    def group_est(self, rep):
        self.group_ls.append(rep)
        print(f"스터디 그룹 {self.group_ls[-1]}이(가) 새롭게 생성되었습니다.")

        if self.parti_ls:
            print("참여자 명단>")
            for p in self.parti_ls:
                print("\t", p[0])
        else:
            pass
````

````python
ml_gstudy = GroupStudy()
````

````python
ml_gstudy.group_est("Supervised Learning")
````

### 예외처리
    - Try, Except 구문을 활용하여 예외처리 시행
    - 코드가 예외발생시 정상 종료가 되도록 설정하여 사용

````python
# 기본형식 예외처리
pw = input("비밀번호를 입력하세요: ")
try:
    print(f'비밀번호의 총 합은 {sum(map(int, pw.split(" ")))}입니다.')
except:
    print("올바른 데이터 타입이 아닙니다. 숫자를 입력하세요.")
````

````python
# try except else - 예외가 발생하지 않으면 else구문이 실행
pw = input("비밀번호를 입력하세요: ")
try:
    res = sum(map(int, pw.split(" ")))
    print(f'비밀번호의 총 합은 {res}입니다.')
except:
    print("올바른 데이터 타입이 아닙니다. 숫자를 입력하세요.")
else:
    if res < 40:
        print("총 합이 조건에 미치지 않으므로 실행할 수 없습니다.")
    else:
        print("Unlocked!")
````

````python
# try except finally - 어떤 처리가 되었든 마지막에 finally 구문 시행
pw = input("비밀번호를 입력하세요: ")
try:
    res = sum(map(int, pw.split(" ")))
    print(f'비밀번호의 총 합은 {res}입니다.')
except:
    print("올바른 데이터 타입이 아닙니다. 숫자를 입력하세요.")
else:
    if res < 40:
        print("총 합이 조건에 미치지 않으므로 실행할 수 없습니다.")
    else:
        print("Unlocked!")
finally:
    print("이용해 주셔서 감사합니다.")
````

#### 사용자 정의 예외처리
    - python에서는 예외가 아니나 사용자가 조건을 설정하여 예외를 발생시켜 비정상종료 시키는 방법
    - 오버라이딩 또는 새로운 클래스 선언을 통해 생성

````python
class LowNumber(Exception):
    def __str__(self):
        return "조건이 맞지 않습니다."
````

````python
def password_input():
    num = sum(map(int, input("조건에 맞는 비밀번호 4자리를 입력하세요: ").split(" ")))
    if num < 50:
        raise LowNumber()
    print(num)
````
